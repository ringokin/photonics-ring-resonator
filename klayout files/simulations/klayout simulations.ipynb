## All ports

```python

#getting the values
intR = lumapi.INTERCONNECT(filename = 'ref all.icp', hide = True)
intR.run()
reference = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()

intR = lumapi.INTERCONNECT(filename = 'simu all.icp', hide = True)
intR.run()
simuPass = intR.getresult('ONA_1','input 1/mode 1/gain')
simuDrop = intR.getresult('ONA_1','input 2/mode 1/gain')
# simuAdd  = intR.getresult('ONA_1','input 3/mode 1/gain')
intR.close()


resultPass = simuPass['TE gain (dB)'] - reference['TE gain (dB)']
resultDrop = simuDrop['TE gain (dB)'] - reference['TE gain (dB)']
# resultAdd  =  simuAdd['TE gain (dB)'] - reference['TE gain (dB)']
```

```python

#plotting
fig1, graph = plt.subplots(1, 2, figsize = (16, 8))
fig1.suptitle("Cumulative Pass", size = 32)

graph[0].plot(simuPass['wavelength'],resultPass)
graph[0].plot(simuDrop['wavelength'],resultDrop)
# graph[0].plot( simuAdd['wavelength'], resultAdd)
graph[0].set_xlim(1.5e-6,1.6e-6)
graph[0].set_xlabel(r'Comprimento de onda',fontsize =15)
graph[0].set_ylabel(r'Transmissão (dB)',fontsize =15)
# graph[0].legend(['Pass','drop', 'add'])
graph[0].legend(['Pass','drop'])

graph[1].plot(simuPass['wavelength'],resultPass)
graph[1].plot(simuDrop['wavelength'],resultDrop)
# graph[1].plot( simuAdd['wavelength'], resultAdd)
graph[1].set_xlim(1.54e-6,1.5475e-6)
graph[1].set_xlabel(r'Comprimento de onda',fontsize =15)
graph[1].set_ylabel(r'Transmissão (dB)',fontsize =15)
# graph[1].legend(['Pass','drop', 'add'])
graph[1].legend(['Pass','drop'])
```

## Pass simulation

```python

#getting the values
intR = lumapi.INTERCONNECT(filename = 'ref pass.icp', hide = True)
intR.run()
referencePass = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()

intR = lumapi.INTERCONNECT(filename = 'simu pass 2x.icp', hide = True)
intR.run()
simuPass2x = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()

intR = lumapi.INTERCONNECT(filename = 'simu pass 3x.icp', hide = True)
intR.run()
simuPass3x = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()

intR = lumapi.INTERCONNECT(filename = 'simu pass 4x.icp', hide = True)
intR.run()
simuPass4x = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()



resultPass2x = simuPass2x['TE gain (dB)'] - referencePass['TE gain (dB)']
resultPass3x = simuPass3x['TE gain (dB)'] - referencePass['TE gain (dB)']
resultPass4x = simuPass4x['TE gain (dB)'] - referencePass['TE gain (dB)']
```

```python

#plotting

fig1, graph = plt.subplots(1, 2, figsize = (16, 8))
fig1.suptitle("Cumulative Pass", size = 32)

graph[0].plot(  simuPass['wavelength'],  resultPass)
graph[0].plot(simuPass2x['wavelength'],resultPass2x)
graph[0].plot(simuPass3x['wavelength'],resultPass3x)
graph[0].plot(simuPass4x['wavelength'],resultPass4x)
graph[0].set_xlim(1.5e-6,1.6e-6)
graph[0].set_xlabel(r'Comprimento de onda',fontsize =15)
graph[0].set_ylabel(r'Transmissão (dB)',fontsize =15)
graph[0].legend(['Pass','Pass2x','Pass3x','Pass4x'])

graph[1].plot(  simuPass['wavelength'],  resultPass)
graph[1].plot(simuPass2x['wavelength'],resultPass2x)
graph[1].plot(simuPass3x['wavelength'],resultPass3x)
graph[1].plot(simuPass4x['wavelength'],resultPass4x)
graph[1].set_xlim(1.54e-6,1.5465e-6)
graph[1].set_xlabel(r'Comprimento de onda',fontsize =15)
graph[1].set_ylabel(r'Transmissão (dB)',fontsize =15)
graph[1].legend(['Pass','Pass2x','Pass3x','Pass4x'])
```

## Drop simulation

```python

#getting the values
intR = lumapi.INTERCONNECT(filename = 'ref drop.icp', hide = True)
intR.run()
referenceDrop = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()

intR = lumapi.INTERCONNECT(filename = 'simu drop 2x.icp', hide = True)
intR.run()
simuDrop2x = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()

intR = lumapi.INTERCONNECT(filename = 'simu drop 3x.icp', hide = True)
intR.run()
simuDrop3x = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()

intR = lumapi.INTERCONNECT(filename = 'simu drop 4x.icp', hide = True)
intR.run()
simuDrop4x = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()



resultDrop2x = simuDrop2x['TE gain (dB)'] - referenceDrop['TE gain (dB)']
resultDrop3x = simuDrop3x['TE gain (dB)'] - referenceDrop['TE gain (dB)']
resultDrop4x = simuDrop4x['TE gain (dB)'] - referenceDrop['TE gain (dB)']
```

```python

#plotting

fig1, graph = plt.subplots(1, 2, figsize = (16, 8))
fig1.suptitle("Cumulative Drop", size = 32)

graph[0].plot(  simuDrop['wavelength'],  resultDrop)
graph[0].plot(simuDrop2x['wavelength'],resultDrop2x)
graph[0].plot(simuDrop3x['wavelength'],resultDrop3x)
graph[0].plot(simuDrop4x['wavelength'],resultDrop4x)
graph[0].set_xlim(1.5e-6,1.6e-6)
graph[0].set_xlabel(r'Comprimento de onda',fontsize =15)
graph[0].set_ylabel(r'Transmissão (dB)',fontsize =15)
graph[0].legend(['Drop','Drop2x','Drop3x','Drop4x'])

graph[1].plot(  simuDrop['wavelength'],  resultDrop)
graph[1].plot(simuDrop2x['wavelength'],resultDrop2x)
graph[1].plot(simuDrop3x['wavelength'],resultDrop3x)
graph[1].plot(simuDrop4x['wavelength'],resultDrop4x)
graph[1].set_xlim(1.5405e-6,1.5465e-6)
graph[1].set_ylim(-10,0)
graph[1].set_xlabel(r'Comprimento de onda',fontsize =15)
graph[1].set_ylabel(r'Transmissão (dB)',fontsize =15)
graph[1].legend(['Drop','Drop2x','Drop3x','Drop4x'])
```

## Ring length simulation

```python

#getting the values
intR = lumapi.INTERCONNECT(filename = 'ref Len.icp', hide = True)
intR.run()
referenceLen = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()

intR = lumapi.INTERCONNECT(filename = 'simu Len 5.icp', hide = True)
intR.run()
simuLen5 = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()

intR = lumapi.INTERCONNECT(filename = 'simu Len 10.icp', hide = True)
intR.run()
simuLen10 = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()

intR = lumapi.INTERCONNECT(filename = 'simu Len 20.icp', hide = True)
intR.run()
simuLen20 = intR.getresult('ONA_1','input 1/mode 1/gain')
intR.close()



resultLen5 = simuLen5['TE gain (dB)'] - referenceLen['TE gain (dB)']
resultLen10 = simuLen10['TE gain (dB)'] - referenceLen['TE gain (dB)']
resultLen20 = simuLen20['TE gain (dB)'] - referenceLen['TE gain (dB)']
```

```python

#plotting

fig1, graph = plt.subplots(1, 2, figsize = (16, 8))
fig1.suptitle("Ring length variation", size = 32)

graph[0].plot(  simuPass['wavelength'],  resultPass)
graph[0].plot(simuLen5['wavelength'],resultLen5)
graph[0].plot(simuLen10['wavelength'],resultLen10)
graph[0].plot(simuLen20['wavelength'],resultLen20)
graph[0].set_xlim(1.5e-6,1.6e-6)
graph[0].set_xlabel(r'Comprimento de onda',fontsize =15)
graph[0].set_ylabel(r'Transmissão (dB)',fontsize =15)
graph[0].legend(['Len','Len + 5um','Len + 10um','Len + 20 $\mu m$'])

graph[1].plot(  simuPass['wavelength'],  resultPass)
graph[1].plot(simuLen5['wavelength'],resultLen5)
graph[1].plot(simuLen10['wavelength'],resultLen10)
graph[1].plot(simuLen20['wavelength'],resultLen20)
graph[1].set_xlim(1.5385e-6,1.546e-6)
graph[1].set_ylim(-5,0)
graph[1].set_xlabel(r'Comprimento de onda',fontsize =15)
graph[1].set_ylabel(r'Transmissão (dB)',fontsize =15)
graph[1].legend(['Len','Len + 5um','Len + 10um','Len + 20 $\mu m$'])
```
